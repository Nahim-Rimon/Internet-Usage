### Libraries Import

```python
import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as figure
import seaborn as sns
```

### Data Import

```python
internet = pd.read_csv(r'C:\Users\RimoNahiM\Documents\Work\CSV & Excel Files\Excel and CSVs\InternetData.csv')
internet.head(10)
```

### Data Manipulation

```python
list(internet.columns)
```

```python
internet.rename(columns={'Avg \n(Mbit/s)Ookla':'Avg Speed'}, inplace=True)
print(internet.shape)
internet.head()
```

```python
print(internet.isnull().sum())
```

```python
internet = internet.fillna(0)
print(internet.shape)
print(internet.isnull().sum())
```

```python
internet.drop_duplicates(inplace = True)
```

### Data Analysis

```python
internet.info()
```

```python
internet.describe()
```

```python
# Change the data type

internet['Internet users'] = internet['Internet users'].astype('int64')
internet['Population'] = internet['Population'].astype('int64')
internet.head()
```

```python
# Population Vs Internet Users

plt.rcParams["figure.figsize"] = (8,8)
sns.scatterplot(data=internet, x='Population', y='Internet users')
plt.title('Internet Users vs Population')
plt.xlabel('Internet users')
plt.ylabel('Population')
plt.show()
```

```python
# Price 1gb 2021 vs Price 1gb 2020

plt.rcParams["figure.figsize"] = (8,8)
sns.regplot(data=internet, x="Avg price of 1GB (Start of 2021)", y="Avg price of 1GB (Start of 2020)")
```

```python
# Correlation between Columns

internet_corr = internet.corr(numeric_only=True)
plt.rcParams["figure.figsize"] = (8,8)
sns.heatmap(internet_corr, annot=True, cmap='BrBG')
plt.title('Correlation Matrix of Internet Dataset')
```

```python
# Which continents has the hihgest avg speed internet?

plt.rcParams["figure.figsize"] = (8,8)
sns.boxplot(data=internet, x="Avg Speed", y="Continental region", palette='Spectral')
```

```python
# Cheapest Continents

plt.rcParams["figure.figsize"] = (8,8)
sns.barplot(data=internet, x='Cheapest 1GB for 30 days (USD)', y='Continental region', palette='Spectral', errorbar=None)
```

```python
# Expensive Continents

plt.rcParams["figure.figsize"] = (8,8)
sns.barplot(data=internet, x='Most expensive 1GB (USD)', y='Continental region', palette='Spectral', errorbar=None)
```
